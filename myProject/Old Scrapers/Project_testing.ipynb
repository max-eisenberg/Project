```python
stats = pd.read_csv("strava_test.csv")
```

```python
stats
```

```python
for i in range(len(stats["Monthly Distance"])):
    stats["Monthly Distance"][i] = int(stats["Monthly Distance"][i])
    stats["Monthly Time"][i] = int(stats["Monthly Time"][i])
    stats["Monthly Elevation"][i] = int(stats["Monthly Elevation"][i].replace(',', ''))
stats
```

```python
dist_tot = 0
time_tot = 0
elev_tot = 0
for i in range(len(stats)):
    dist_tot += stats["Monthly Distance"][i]
    time_tot += stats["Monthly Time"][i]
    elev_tot += stats["Monthly Elevation"][i]

ave_dist = dist_tot/len(stats)
ave_time = time_tot/len(stats)
ave_elev = elev_tot/len(stats)
```

```python
print("Looking at your competitors, this is their average activity for the past week:\n")
print(f"Average distance per Week: {round(ave_dist,2)} miles.")
print(f"Average hours per week: {round(ave_time,2)} hours.")
print(f"Average elevation per week: {round(ave_elev,2)} feet.")
```

```python
dist_tot = 0
time_tot = 0
elev_tot = 0
counter = 0
for i in range(len(stats2)):
    if stats2["Monthly Distance"][i] != 0.0:
        dist_tot += stats2["Monthly Distance"][i]
        time_tot += stats2["Monthly Time"][i]
        elev_tot += stats2["Monthly Elevation"][i]
        counter += 1

ave_dist = dist_tot/counter
ave_time = time_tot/counter
ave_elev = elev_tot/counter
```
